```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv("StudentPerformanceFactors.csv")
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python

avg_score_by_parental_involvement = df.groupby('Parental_Involvement')['Exam_Score'].mean()

avg_score_by_parental_involvement
```

```python
avg_score_by_resources = df.groupby('Access_to_Resources')['Exam_Score'].mean()
avg_score_by_resources
```

```python
avg_score_by_extracurricular = df.groupby('Extracurricular_Activities')['Exam_Score'].mean()

avg_score_by_extracurricular
```

```python
avg_score_by_motivation = df.groupby('Motivation_Level')['Exam_Score'].mean()
avg_score_by_motivation
```

```python
avg_score_by_income = df.groupby('Family_Income')['Exam_Score'].mean()
avg_score_by_income
```

```python
avg_score_by_gender = df.groupby('Gender')['Exam_Score'].mean()
avg_score_by_gender
```

```python
avg_score_by_distance = df.groupby('Distance_from_Home')['Exam_Score'].mean()
avg_score_by_distance
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Hours_Studied', y='Exam_Score', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title("Hours Studied vs Exam Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='Attendance', y='Exam_Score', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'blue'})
plt.title("Attendance vs Exam Scores")
plt.xlabel("Attendance (%)")
plt.ylabel("Exam Score")
plt.show()
```
